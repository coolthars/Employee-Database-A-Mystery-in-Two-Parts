```python
# Dependencies and Setup
# SQL Alchemy
from sqlalchemy import create_engine

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# NumPy
import numpy as np
```

```python
# Create Engine and Pass in Postgres Connection
# Setup to Connect to Database 
engine = create_engine('postgres://postgres:password$@localhost:5432/test')
conn = engine.connect()
```

```python
# Query All Records in the Salaries Table
salaries_data = pd.read_sql("SELECT * FROM salaries", conn)
salaries_data.head()
```

```python
# Query All Records in the Titles Table
titles_data = pd.read_sql("SELECT * FROM titles", conn)
titles_data.head()
```

```python
# Merge Columns / Combine Data
combined_data = pd.merge(salaries_data, titles_data, on="emp_no", how="inner")
combined_data.head()
```

```python
# Groupby Title
group_data=combined_data.groupby("title").mean()
group_data.head()
```

```python
# Drop emp_no from grouped_df to Show Only title & salary
final_group_data= group_data.drop(columns = "emp_no")
final_group_data
```

```python
# Reset Index
final_group_data = final_group_data.reset_index()
final_group_data
```

```python
# Set x_axis, y_axis & Tick Locations
x_axis = final_group_data["title"]
ticks = np.arange(len(x_axis))
y_axis = final_group_data["salary"]
 
# Create Bar Chart Based on Above Data
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=['r','g','b','c','y','m','k'])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Titles")
plt.title("Average  Salary by Title")

# Save Figure
# plt.savefig("./Images/average_salary_by_title.png")
plt.savefig("salary_by_title.png",bbox_inches = "tight")

# Show plot
plt.show()
```
